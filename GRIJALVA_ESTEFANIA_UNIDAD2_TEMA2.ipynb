```python
class NodoPedido:
    def __init__(self, id_pedido: int, tipo: str, descripcion: str):
        self.id_pedido = id_pedido 
        self.tipo = tipo 
        self.descripcion = descripcion 
        self.siguiente = None 

class GestorPedidos:
    def __init__(self):
        self.cola_frente = None
        self.cola_fin = None
        self.pila_tope = None

    def registrar_pedido(self, id_pedido: int, tipo: str, descripcion: str) -> None:
        nuevo_nodo = NodoPedido(id_pedido, tipo, descripcion)
        if tipo == "ESTÁNDAR":
            if self.cola_frente is None:
                self.cola_frente = nuevo_nodo
                self.cola_fin = nuevo_nodo
            else:
                self.cola_fin.siguiente = nuevo_nodo
                self.cola_fin = nuevo_nodo
            print(f"[REGISTRO] Pedido Estándar #{id_pedido} encolado exitosamente.")
        elif tipo == "RECLAMO":
            if self.pila_tope is None:
                self.pila_tope = nuevo_nodo
            else:
                nuevo_nodo.siguiente = self.pila_tope
                self.pila_tope = nuevo_nodo
            print(f"[REGISTRO] Reclamo Urgente #{id_pedido} apilado en el tope.")

    def despachar_siguiente(self) -> NodoPedido:

        if self.pila_tope is not None:
            nodo_despachado = self.pila_tope 
            self.pila_tope = self.pila_tope.siguiente
            nodo_despachado.siguiente = None 
            return nodo_despachado
        elif self.cola_frente is not None:
            nodo_despachado = self.cola_frente
            self.cola_frente = self.cola_frente.siguiente
            if self.cola_frente is None:
                self.cola_fin = None
            nodo_despachado.siguiente = None
            return nodo_despachado
        else:
            return None

    def activar_contingencia_masiva(self) -> None:

        if self.cola_frente is None:
            print("[CONTINGENCIA] La cola estándar está vacía. No hay elementos que trasladar.")
            return
        print("[ALERTA] Hora Pico detectada. Activando Modo de Contingencia Masiva...")
        while self.cola_frente is not None:
            nodo_actual = self.cola_frente
            self.cola_frente = self.cola_frente.siguiente
            nodo_actual.siguiente = self.pila_tope
            self.pila_tope = nodo_actual
        self.cola_fin = None
        print("[SISTEMA] Contingencia finalizada con éxito. Cola congelada y vaciada.")

    def mostrar_estado_sistema(self) -> None:
        print(" ESTADO ACTUAL DEL SISTEMA ")
        print()
        print("PILA DE RECLAMOS (Tope -> Fin):")
        aux = self.pila_tope
        if aux is None: print("  [Vacía]")
        while aux is not None:
            print(f"  -> [Pedido #{aux.id_pedido} | {aux.tipo} | '{aux.descripcion}']")
            aux = aux.siguiente
            
        print("COLA ESTÁNDAR (Frente -> Fin):")
        aux = self.cola_frente
        if aux is None: print("  [Vacía]")
        while aux is not None:
            print(f"  -> [Pedido #{aux.id_pedido} | {aux.tipo} | '{aux.descripcion}']")
            aux = aux.siguiente
        print("=" * 85)
if __name__ == "__main__":
    gestor = GestorPedidos()
    gestor.registrar_pedido(101, "ESTÁNDAR", "1 Hamburguesa con papas")
    gestor.registrar_pedido(102, "ESTÁNDAR", "1 Pizza Familiar Pepperoni")
    gestor.registrar_pedido(103, "ESTÁNDAR", "2 Combos de Tacos")
    gestor.registrar_pedido(501, "RECLAMO", "Mi pedido llegó frío")
    gestor.registrar_pedido(502, "RECLAMO", "Falta la bebida en el combo")
    gestor.mostrar_estado_sistema()
    p1 = gestor.despachar_siguiente()
    print(f"[DESPACHO] Atendido de inmediato: #{p1.id_pedido} ({p1.tipo}) - Motivo: {p1.descripcion}")
    print()
    gestor.mostrar_estado_sistema()
    gestor.activar_contingencia_masiva()
    gestor.mostrar_estado_sistema()
    print(" Despachando remanentes de contingencia")
    print()
    while True:
        pedido = gestor.despachar_siguiente()
        if pedido is None:
            print("[INFO] No quedan más pedidos por despachar en el sistema.")
            break
        print(f"[DESPACHO CONTINGENCIA] Atendido: #{pedido.id_pedido} - Contenido: {pedido.descripcion}")
```
